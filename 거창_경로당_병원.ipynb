```python
hospital_position = []
for i in range(len(df_hospital.values)):
    hospital_position.append(df_hospital.values[i][2:4])
print(np.array(hospital_position).shape)
hospital_position = np.array(hospital_position)
hospital_position = np.array(hospital_position)[:,[1,0]]
hospital_position = np.array([tuple(row) for row in hospital_position])
print(hospital_position)
```

```python
elderly_position = []
for i in range(len(df_elderly.values)):
    elderly_position.append(df_elderly.values[i][6:8])
print(np.array(elderly_position).shape)
elderly_position = np.array(elderly_position)
elderly_position = np.array([tuple(row) for row in elderly_position])
print(elderly_position)
```

```python
from haversine import haversine_vector
distance = haversine_vector(elderly_position, hospital_position, comb=True, unit = 'km')
```

```python
distance.shape
```

```python
def indices_of_less_than_five(matrix):
    result = []
    for i, row in enumerate(matrix):
        for j, value in enumerate(row):
            if value < 3:
                result.append((i, j))
    return result
```

```python
filter = indices_of_less_than_five(distance)
```

```python
print(filter)
```

```python
print(np.array(filter).shape)
```

```python
senior = []
hosp = []
for idx in filter:
    row_idx, col_idx = idx
    senior.append(senior_center[col_idx])
    hosp.append(hospital[row_idx])
print(senior)
print(hosp)
```

```python
senior = np.array(senior)
hosp = np.array(hosp)
```

```python
senhos = np.concatenate((senior.reshape(-1,1), hosp.reshape(-1,1)), axis=1)
print(senhos)
```

```python
print(senhos.shape)
```

```python
from collections import defaultdict
hospital_dict = defaultdict(list)
for item in senhos:
    hospital_dict[item[0]].append(item[1])
hospital_dict = dict(hospital_dict)
hospital_dict
#print(hospital_dict["대곡경로당"])
```

```python
for key, value in hospital_dict.items():
    print(f"{key}의 병원 개수:", len(value))
```

```python
position_elderly = [
                    {"name":senior_center[i],
                    "lat": elderly_position[i][0],
                    "lon": elderly_position[i][1]}
                    for i in range(len(senior_center))
]
position_hospital = [
                    {"name":hospital[i],
                    "lat": hospital_position[i][0],
                    "lon": hospital_position[i][1]}
                    for i in range(len(hospital))
]
```

```python
import folium
from geopy.distance import geodesic


radius_km = 3

m = folium.Map(location=[position_elderly[0]["lat"], position_elderly[0]["lon"]], zoom_start=13)

for loc in position_elderly:
    folium.Marker([loc["lat"], loc["lon"]], popup=loc["name"]).add_to(m)

    folium.Circle(
            radius = radius_km * 1000,
            location=[loc["lat"], loc["lon"]],
            popup=f"{loc['name']}",
            color="blue",
            fill=True,
            fill_color="blue",
            fill_opacity=0.2
    ).add_to(m)

m.save("map.html")

m
```

```python

radius_km = 3

search_elderly = input("경로당:")

m = folium.Map(location=[position_elderly[0]["lat"], position_elderly[0]["lon"]], zoom_start=13)

for loc in position_elderly:
    if loc["name"].lower() == search_elderly.lower():
        folium.Circle(
                radius = radius_km * 1000,
                location=[loc["lat"], loc["lon"]],
                popup=f"{loc['name']}",
                color="blue",
                fill=True,
                fill_color="blue",
                fill_opacity=0.2
        ).add_to(m)
        folium.Marker([loc["lat"], loc["lon"]], popup=loc["name"]).add_to(m)
        m.location = [loc["lat"], loc["lon"]]
        m.zoom_start = 12
        if loc["name"] in hospital_dict.keys():
            for lis in hospital_dict[loc["name"]]:
                for pos in position_hospital:
                    if pos["name"] == lis:
                        folium.Marker([pos["lat"], pos["lon"]], popup=pos["name"], icon=folium.Icon(icon="plus", color="red")).add_to(m)
        break

m.save("map.html")
m
```

```python

radius_km = 3


m = folium.Map(location=[position_elderly[0]["lat"], position_elderly[0]["lon"]], zoom_start=13)
elderly_count = 0
hospital_count = set()
for loc in position_elderly:
    if (loc["name"] not in hospital_dict.keys()) or (len(hospital_dict[loc["name"]]) <= 5):
        folium.Circle(
                radius = radius_km * 1000,
                location=[loc["lat"], loc["lon"]],
                popup=f"{loc['name']}",
                color="blue",
                fill=True,
                fill_color="blue",
                fill_opacity=0.2
        ).add_to(m)
        folium.Marker([loc["lat"], loc["lon"]], popup=loc["name"]).add_to(m)
        m.location = [loc["lat"], loc["lon"]]
        m.zoom_start = 5
        elderly_count += 1
        if loc["name"] in hospital_dict.keys():
            for lis in hospital_dict[loc["name"]]:
                for pos in position_hospital:
                    if pos["name"] == lis:
                        hospital_count.add(lis)
                        folium.Marker([pos["lat"], pos["lon"]], popup=pos["name"], icon=folium.Icon(icon="plus", color="red")).add_to(m)
print(f"소외 경로당 개수: {elderly_count}")
print(f"병원 개수: {len(hospital_count)}")
print(hospital_count)
m.save("map.html")
m
```
