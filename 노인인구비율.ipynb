```python
from google.colab import drive
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import shutil


drive.mount('/content/drive')

font_path = '/content/drive/MyDrive/MALGUN.TTF'
shutil.copy(font_path, '/usr/share/fonts/truetype/MALGUN.TTF')  #  Colab  

#   
fm._load_fontmanager(try_read_cache=False)

fontprop = fm.FontProperties(fname='/usr/share/fonts/truetype/MALGUN.TTF')
plt.rc('font', family=fontprop.get_name())

data = pd.read_csv('/content/drive/MyDrive/_.csv', encoding='EUC-KR')

plt.figure(figsize=(10, 6))
bars = plt.bar(data[''], data[''], color='navy')
plt.ylim(20, 60)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.1f}%', ha='center', fontproperties=fontprop)

plt.xlabel('', fontproperties=fontprop)
plt.ylabel('(%)', fontproperties=fontprop)
plt.title('  ', fontproperties=fontprop)
plt.xticks(rotation=45, fontproperties=fontprop)
plt.tight_layout()
plt.show()
```
