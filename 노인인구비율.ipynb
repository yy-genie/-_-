```python
from google.colab import drive
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd
import shutil


drive.mount('/content/drive')

font_path = '/content/drive/MyDrive/MALGUN.TTF'
shutil.copy(font_path, '/usr/share/fonts/truetype/MALGUN.TTF')  # 폰트를 Colab 환경으로 복사

# 폰트 캐시 초기화
fm._load_fontmanager(try_read_cache=False)

fontprop = fm.FontProperties(fname='/usr/share/fonts/truetype/MALGUN.TTF')
plt.rc('font', family=fontprop.get_name())

data = pd.read_csv('/content/drive/MyDrive/거창_지도.csv', encoding='EUC-KR')

plt.figure(figsize=(10, 6))
bars = plt.bar(data['행정구역'], data['노인인구비율'], color='navy')
plt.ylim(20, 60)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.1f}%', ha='center', fontproperties=fontprop)

plt.xlabel('행정구역', fontproperties=fontprop)
plt.ylabel('노인인구비율(%)', fontproperties=fontprop)
plt.title('거창군 행정구역별 노인인구비율', fontproperties=fontprop)
plt.xticks(rotation=45, fontproperties=fontprop)
plt.tight_layout()
plt.show()
```
